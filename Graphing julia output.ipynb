```python
X_TICKS = 100 # use every 4th value; change this number until you like the result

times=df.times[0:500]
predictions=df.da_price[0:500]
rvpredictions=df.SOC[0:500]*100-120
plt.title('SOC and DA prices')
plt.plot(predictions, '--', color='darkred', label='DA Prices')
plt.plot(rvpredictions, '--', color='darkorange', label='SOC')
plt.legend()
plt.ylim(-120, 120)
plt.xticks(range(0, len(times), X_TICKS), times[::X_TICKS], rotation = 90)
plt.tight_layout() # auto adjust the spacing around the graph
plt.show()
```

```python
profit = 0
for i in range(3*48):
    profit_da = df.da_price[i]*(1/2)*(df.p_da_dis[i]-df.p_da_ch[i])
    profit_id = df.id_prices[i]*(1/2)*(df.p_id_dis[i]-df.p_id_ch[i])
    profit = profit + profit_da + profit_id
print(profit)
```
